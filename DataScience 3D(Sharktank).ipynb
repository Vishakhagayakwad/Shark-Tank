# 5.	Find Interest of Sharks 

```python
shark_in = [df1['Anupam'].sum(),df1['Ashneer'].sum(),df1['Namita'].sum(),df1['Aman'].sum(),df1['Peyush'].sum(),df1['Vineeta'].sum(),df1['Ghazal'].sum()]
```

```python
sharks = ['Anupam','Ashneer','Namita','Aman','Peyush','Vineeta','Ghazal']
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x=sharks, y=shark_in, palette='rainbow')
plt.xlabel('Sharks')
plt.ylabel('Number of Companies they were interested In')
plt.title("Interest of Sharks", fontsize = 15)
plt.plot()
```

# 7.	Find Top 5 Investments 

```python
investment=df.groupby('Brand')['Investment Amount (In Lakhs INR) '].sum().sort_values(ascending = False).reset_index().head(5)
```

```python
plt.figure(figsize=(8,5))
sns.barplot(x=investment['Brand'], y=investment['Investment Amount (In Lakhs INR) '],palette='rocket')
plt.xticks(rotation = 45)
plt.xlabel('Brands')
plt.ylabel('Investment Amount (In Lakhs INR)')
plt.title("Top 5 Investments", fontsize = 15)
plt.plot()
```

# 8.	Find Top 10 Valued Companies 

```python
top_valued=df.groupby('Brand')['Valuation Offered (In Lakhs INR)'].sum().sort_values(ascending = False).reset_index().head(10)
```

```python
plt.figure(figsize=(8,5))
sns.barplot(x=top_valued['Brand'], y=top_valued['Valuation Offered (In Lakhs INR)'],palette='rocket')
plt.xticks(rotation = 45)
plt.xlabel('Brands')
plt.ylabel('Valuation Offered (In Lakhs INR)')
plt.title("Top 10 valued companies", fontsize = 15)
plt.plot()
```
